```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
df.melt(id_vars=None, value_vars=None, var_name=None, value_name="value", col_level=None)
```

```python
df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
melted_df = df.melt(
    id_vars=["Name"],
    value_vars=["Math", "Science", "English"],
    var_name="Subject",
    value_name="Score"
)
```

```python
melted_df
```

```python
pivot_df = melted_df.pivot(
    index="Name",
    columns="Subject",
    values="Score"
)

pivot_df
```
